```python
"""
Mini Challenge: Number Guess Counter

Write a Python program that:

Store a secret number:

secret = 7
Ask the user to guess the number.
Keep asking until the user enters the correct number.
Count how many attempts the user took.

When they guess correctly, print:

Correct!
You guessed it in X attempts.
Example:
Guess the number: 3
Wrong! Try again.
Guess the number: 10
Wrong! Try again.
Guess the number: 7
Correct!
You guessed it in 3 attempts"""
secert = 7
guess = int(input("guess the number: "))
count = 1

while guess != secert:
   print("wrong! Try again.")
   guess = int(input("guess the number: "))
   count += 1
   print(f"no of attempts :{count}")

if guess == secert:
  print("correct!")
print(f"You guessed it in {count} attempts")
```
