```python
#Print all even numbers from 1 to n
n = int(input("enter ur limit: "))
i = 1
while i <= n:
  if i % 2 == 0:
    print(i)
  i += 1
```

```python
#Print all odd numbers from 1 to n.
n = int(input("enter ur limit: "))
i=1
while i <= n:
  if i % 2 != 0:
    print(i)
  i += 1
```
