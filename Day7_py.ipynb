```python
#Write a function that takes two numbers and returns their sum.
def twonumbers(a,b):
  total = sum([a,b])
  return(total)
twonumbers(5,7)
```

```python
#Write a function that returns the square of a number
def square(a):
  return a**2
square(9)
```

```python
#Write a function that checks whether a number is even or odd.
def checknumber(a):
  if(a % 2 == 0):
    return("even")
  else:
    return("odd")
checknumber(3)
```

```python
#Write a function that returns the larger of two numbers.
def largernumber(a,b):
  if(a>b):
   return(a)
  elif(a<b):
   return(b)
  else:
   return(a)
largernumber(8,8)
```
