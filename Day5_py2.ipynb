```python
"""Take a student's fee amount.
Print "Payment Pending"
If the amount is greater than 0,
Otherwise print "Fees Cleared"."""
Stu_fee = int(input("enter ur fee amount"))
if(Stu_fee == 0):
  print("fees cleared ")
elif(Stu_fee > 0):
  print("Payment Pending")
else:
  print("please enter correct fee!")
```

```python

"""Take a roll number
 And print only the last 4 characters
 using slicing."""
Roll_num = input("enter ur roll number")
print(Roll_num[-4:])
```

```python
"""Take a student's percentage. Print:
>= 90 → "Excellent"
75–89 → "Good"
50–74 → "Average"
< 50 → "Needs Improvement" """
Stu_per = int(input("enter ur percentage"))
if Stu_per >= 90:
  print("excellent")
elif Stu_per >= 75 and Stu_per <= 89:
  print("good")
elif Stu_per >= 50 and Stu_per <= 74:
  print("Average")
elif Stu_per < 50:
  print("needs improvement")
```

```python

"""Take a payment status
(paid or pending)
 And print an appropriate message
 using if-else."""
payment_status = input("enter ur payment status ")
if(payment_status == "paid"):
  print("fee cleared")
else:
  print("payment pending")
```
