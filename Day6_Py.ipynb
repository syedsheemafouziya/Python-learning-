```python
"""Take a list of 10 numbers and print:
Largest number
Smallest number
Sum of all numbers"""
list_num = [10,50,17,4,84,39,22,78,23,20]
print(list_num)
Largest_num = max(list_num)
Smallest_num = min(list_num)
total = sum(list_num)
print(f" the largest number {Largest_num}\n the smallest number{Smallest_num}\n the Sum of all numbers {total}")
```

```python
"""Create a list of 10 numbers and print:
First element
Last element
Length of the list"""
list_num = [10,50,17,4,84,39,22,78,23,20]
print(list_num[0])
print(list_num[-1])
print(len(list_num))
```

```python
#Check whether 50 is present in the list.
list_num = [10,50,17,4,84,39,22,78,23,20]
print(50 in list_num)
```

```python
#Check whether 50 is present in the list.
list_num = [10,50,17,4,84,39,22,78,23,20]
if 50 in list_num:
  print("50 present in list")
else:
  print("50 is not present in list")
```

```python
#Sort a list in ascending order
list_num = [10,50,17,4,84,39,22,78,23,20]
list_num.sort()
print(list_num)
```

```python
#Reverse a list.
list_num = [10,50,17,4,84,39,22,78,23,20]
print(list_num[::-1])
```

```python
"""Count how many times 10 appears in this list:
[10, 20, 10, 30, 40, 10, 50]"""
list_num = [10, 20, 10, 30, 40, 10, 50]
print(list_num.count(10))
```

```python
#Find the index of 84 in a list.
list_num = [10,50,17,4,84,39,22,78,23,20]
print(list_num.index(84))
```

```python
"""Create two sets and print:
Union
Intersection
Difference"""
set1 = {12,44,"apple",88,66,43,90}
set2 = {90,58,67,43,22,89,"dog"}
print(set1 | set2)
print(set1 & set2)
print(set1 - set2)
```
